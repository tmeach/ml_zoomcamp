# Multiply matrix X matrix

### Vector - vector multiplication (  )

```python
b = np.array(b)
b
```

```python
c = np.array(c)
c
```

```python
def vector_vector_multy(b, c):
    assert b.shape[0] == c.shape[0]
    
    n = b.shape[0]
    
    result = 0.0
    
    for i in range(n):
        result = result + b[i] * c[i]
    
    return result
```

```python
vector_vector_multy(b, c)
```

### Actually in Numpy we already have dot product formula and it's not neccesary to write function:

```python
b.dot(c)
```

### Matrix - vector multiplication

### Translate formula into Py code

```python
U = np.array([
    [2, 4, 5, 6],
    [1, 2, 1, 2],
    [3, 1, 2, 1]
])
```

```python
U
```

```python
def matrix_vector_mult(U, b):
    assert U.shape[1] == b.shape[0]
    
    num_rows = U.shape[0]
    
    result = np.zeros(num_rows)
    
    for i in range(num_rows):
        result[i] = vector_vector_multy(U[i], b)
        
    return result
```

### Our result 

```python
matrix_vector_mult(U, b)
```

### Again instead of using py function in Numpy we can just use dot method

```python
U.dot(b)
```

We get same result 

### Matrix - matrix multiplication

### Implement this into py code

```python
U
```

```python
V = np.array([
    [1, 1, 2],
    [0, 0.5, 1],
    [0, 2, 1],
    [2, 1, 0]
])
```

```python
V
```

```python
def matrix_matrix_mult(U, V): 
    assert U.shape[1] == V.shape[0]
    
    num_rows = U.shape[0]
    num_col = V.shape[1]
    
    result = np.zeros((num_rows, num_col))
    
    for i in range(num_col):
        vi = V[:, i]
        Uvi = matrix_vector_mult(U, vi)
        result[:, i] = Uvi
        
    return result
```

```python
matrix_matrix_mult(U, V)
```

### Again NumPy can do the same operation without py function

```python
U.dot(V)
```
